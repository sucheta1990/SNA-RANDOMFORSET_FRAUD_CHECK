```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# for fraud_check using abbrevations f_c
f_c=pd.read_csv("Fraud_check (1).csv")
f_c.head(3)
```

```python
f_c.rename(columns={"Marital.Status":"MartialStatus",
                    "Taxable.Income":"TaxableIncome",
                    "City.Population":"CityPopulation",
                    "Work.Experience":"WorkExperience"},inplace=True)
```

```python
f_c.head(3)
```

```python
# lets create the new column called tax_range bsed on the taxable income greter than 30000 as good and below that is risky
```

```python
tax_range=[]
for tax in f_c["TaxableIncome"]:
    if tax >=30000:
        tax_range.append("Good")
    elif tax <0 and tax> 99620:
        tax_range.append("NaN")
    else:
        tax_range.append("Risky")
        print(f_c)
    f_c["tax_range"]=pd.Series(tax_range)
```

LETS USE LABEL ENCODING TO CONVERT THE SOME OF THE FEATURE COLUMNS INTO LABELS

```python
from sklearn.preprocessing import LabelEncoder
lm=LabelEncoder()
f_c["Undergrad"]=lm.fit_transform(f_c["Undergrad"])
f_c["MartialStatus"]=lm.fit_transform(f_c["MartialStatus"])
f_c["Urban"]=lm.fit_transform(f_c["Urban"])
f_c["tax_range"]=lm.fit_transform(f_c["tax_range"])
```

LETS STANDARDISE THE FEATURE COLUMNS AND NORMALISE TO THE 0,1 SCALE

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
f_c_norm=norm_fun(f_c)
```

```python
f_c_norm.head(3)
```

 LETS ASSUME THAT IF THE TAXABLE INCOME<= 30000 THEN ITS CONSIDERED AS RISKY=1,AND IF TAXABLE INCOME IS ABOVE 30000 
 ITS CONSIDERED AS GOOD=0

```python
x=f_c_norm.iloc[:,0:5]
y=f_c_norm.iloc[:,-1]
```

```python
# lets divide the componenets into train and test components
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=10)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
pd.Series(y_pred).value_counts()
```

```python
np.mean(y_pred==y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
accuracy_score(y_test,y_pred)
```
